```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]','',text)
    txt = []
    for w in text.split():
        stemWord = porter.stem(w)
        txt.append(stemWord)
    txt = ' '.join(txt)    
    return txt
```

```python
#clean text 
df['Text'] = df['Text'].apply(clean_text)
df.head() 
```

```python
def extract_gram(filtter,num):
        filtter  = " ".join(filtter)
        n_grams = ngrams(nltk.word_tokenize(filtter), num)
        return [" ".join(grams) for grams in n_grams]
#print("1-gram :" , extract_gram(filtter , 1))   
print("2-gram :" , extract_gram(filtter , 2))   
#print("3-gram :" , extract_gram(filtter , 3))   
#print("4-gram :" , extract_gram(filtter ,4))     
```

```python
def Count(text):
    fd =nltk.FreqDist(text)
    return fd
```

```python
biGramCount = Count(extract_gram(filtter  , 2))
uniGramCount = Count(extract_gram(filtter  , 1))
#biGramCount
#uniGramCount
```

```python
for k,v in biGramCount.items():
    t = k.split()
    countUnigram = uniGramCount[t[0]]
    p = v / countUnigram
    print(f'P({k}|{t[0]}) = {p}')
```
